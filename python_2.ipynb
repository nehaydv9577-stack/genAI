```python
#loop and control stataement
#break -- this statement break the loop
i = 0
while i<5:
  i+=1 #increment
  if i==3:
    break
  else:
      print(i)
```

```python
#continue statement -- this statement skip the current itration
i = 0
while i<5:
  i+=1
  if i==3:
    continue
    print(i)
  else:
    print(i)
```

```python
#pass statement-- this statement pass the code
i = 0
while i<5:
  i+=1
  if i == 3:
    pass
    print(i)
  else:
    print(i)
```

```python
#default argument
def area_perimeter(width=4, height=8):
  area = width*height
  perimeter = 2*(width + height)
  return "Area is",area, "and perimeter is", perimeter
```

```python
print(area_perimeter(10))
```

```python
#keyword argument
def interest(p,r,t):
  i = (p*r*t)/100
  return i
```

```python
print(interest(r=10,t=20,p=1000))
```

```python
#mixed argument
#positional argument follow keyword argument
print(interest(1000,t =2,r=10))#first positional argument an then keyword
```

```python
#variable lenggth argument
#non-keyword positional variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(2,3,4,5,6,9)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:#args= argument
    sum+=num  #sum=sum+num
  return sum
```

```python
sum_number(3,5,6,8,9)
```

```python
def mul_number(*args):
  mul = 1
  for num in args:
    mul/=num
  return mul
```

```python
 mul_number(100,50)
```

```python
# keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(type(kwargs))
  print(len(kwargs))
```

```python
test(a=10,b=20,c=30)
```

```python
# @title Default title text
def pass_students(**kwargs):
  pass_student_list =[]
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_students(A=30, B=40, C=50, D=60, E=70)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fruits = np.array(["apple","banana","orange","watermellon"])
quantity = np.array([100,120,80,50])
color=np.array(["red","yellow","orange","green"])
plt.bar(fruits,quantity,color= color)
plt.xlabel("fruits")
plt.ylabel("quantity")
plt.title("fruit quantity")
plt.show()
```

```python
fruits = np.array(["apple","banana","orange","watermellon"])
quantity = np.array([100,120,80,50])
color=np.array(["red","yellow","orange","green"])
plt.barh(fruits,quantity,color= color)
plt.xlabel("quantity")
plt.ylabel("fruits")
plt.title("fruit quantity")
plt.show()
```

```python
#pie plot
quantity = np.array([20,35,15,30])
fruits = np.array(["apple","banana","orange","mango"])
myexplode=[0.1,0,0,0]#This helps focus attention on one category.
plt.pie(quantity,labels=fruits,explode=myexplode)
plt.legend(title="fruits",loc="best")

plt.show()
```
